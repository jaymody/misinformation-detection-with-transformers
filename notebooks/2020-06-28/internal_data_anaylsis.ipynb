# phase2 claims related articles overlap

- there is overlap in the claims for their related articles set

```python
import json
with open("data/phase2/all-data/metadata.json", "r") as fi:
    metadata = json.load(fi)

arts_vals_list = [v for d in metadata for v in d["related_articles"].values()]
arts_keys_list = [v for d in metadata for v in d["related_articles"].keys()]
arts_vals_set = set(arts_vals_list)
arts_keys_set = set(arts_keys_list)

print(len(arts_vals_list))
print(len(arts_keys_list))
print(len(arts_vals_set))
print(len(arts_keys_set))
```

# phase1 vs phase2
---

```python
import os
import json

from tqdm import tqdm

from valerie.data import load_claims, load_articles, save_claims, save_articles
```

```python
claims_phase1 = load_claims("data/phase1/all-data/claims.json")
articles_phase1 = load_articles("data/phase1/all-data/articles.json")

claims_phase2 = load_claims("data/phase2/all-data/claims.json")
articles_phase2 = load_articles("data/phase2/all-data/articles.json")
```

```python
print(len(claims_phase1), len(claims_phase2))
print(len(articles_phase1), len(articles_phase2))
```

```python
eq1 = []
eq2 = []
eq3 = []
eq4 = []
for k1, claim1 in tqdm(claims_phase1.items()):
    for k2, claim2 in claims_phase2.items():
        if claim1.claim == claim2.claim and claim1.label == claim2.label and claim1.claimant == claim2.claimant:
            eq1.append((k1, k2))
        elif claim1.claim == claim2.claim and claim1.label == claim2.label and claim1.claimant != claim2.claimant:
            eq2.append((k1, k2))
        elif claim1.claim == claim2.claim and claim1.label != claim2.label and claim1.claimant == claim2.claimant:
            eq3.append((k1, k2))
        elif claim1.claim == claim2.claim and claim1.label != claim2.label and claim1.claimant != claim2.claimant:
            eq4.append((k1, k2))
```

```python
print(len(eq1))
print(len(eq2))
print(len(eq3))
print(len(eq4))
print(len(eq1) + len(eq2) + len(eq3) + len(eq4))
```
