```python
import json

from valerie.data import Claim, load_claims
```

```python
claims = load_claims("data/phase2/all-data/claims.json")
```

```python
with open("logs/tests/test-predictions.json") as fi:
    predictions = {int(k): {
        "claim_id": int(k),
        "claim": claims[int(k)].claim,
        "claimant": claims[int(k)].claimant,
        "date": claims[int(k)].date,
        "actual_label": claims[int(k)].label,
        "predicted_label": v["label"],
        "num_correct": sum(1 for art in v["related_articles"] if art in claims[int(k)].related_articles.values()),
        "actual_articles": sorted(list(claims[int(k)].related_articles.values())),
        "predicted_articles": sorted(v["related_articles"]),
        "explanation": v["explanation"],
    } for k, v in json.load(fi).items()}
```

```python
def pretty_prediction(pred, verbose=False):
    terms = []
    if verbose:
        terms.append("claim_id:\t\t" + str(pred["claim_id"]))
        terms.append("actual_label:\t\t" + str(pred["actual_label"]))
        terms.append("predicted_label:\t" + str(pred["predicted_label"]))
        terms.append("num_correct:\t\t" + str(pred["num_correct"]))
        terms.append("actual_articles:" + str(json.dumps(pred["actual_articles"], indent=2)))
        terms.append("predicted_articles:" + str(json.dumps(pred["predicted_articles"], indent=2)))
    terms.append("\nclaimant:\t" + str(pred["claimant"]))
    terms.append("date:\t\t" + str(pred["date"]))
    terms.append("\nclaim:\n" + str(pred["claim"]))
    terms.append("\nexplanation:\n" + str(pred["explanation"]))
    return "\n".join(terms)
```

```python
text = ""
for pred in predictions.values():
    text += pretty_prediction(pred, verbose=False)
    text += "\n\n" + "_"*100 + "\n\n"
with open('logs/pretty_predictions.txt', 'w') as fo:
    fo.write(text)
    
text_verbose = ""
for pred in predictions.values():
    text_verbose += pretty_prediction(pred, verbose=True)
    text_verbose += "\n\n" + "_"*100 + "\n\n"
with open('logs/pretty_predictions_verbose.txt', 'w') as fo:
    fo.write(text_verbose)
```
