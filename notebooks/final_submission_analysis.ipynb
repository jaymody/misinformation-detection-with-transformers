```python
import pandas as pd

out_df = pd.read_json("data/phase2-validation-100/raw/metadata.json")
df7 = pd.read_json("logs/predictions/7_predictions.json", orient="index").sort_index()
df9 = pd.read_json("logs/predictions/9_predictions.json", orient="index").sort_index()
df10 = pd.read_json("logs/predictions/10_predictions.json", orient="index").sort_index()
```

```python
for i, exp in enumerate(df7["explanation"]):
    out_df.loc[i, "explanation7"] = exp

for i, exp in enumerate(df9["explanation"]):
    out_df.loc[i, "explanation9"] = exp

for i, exp in enumerate(df9["explanation"]):
    out_df.loc[i, "explanation10"] = exp

out_df["explanation7_score"] = ""
out_df["explanation9_score"] = ""
out_df["explanation10_score"] = ""
```

```python
for exp_num in range(100):
    print(out_df["claim"][exp_num], "-", out_df["claimant"][exp_num])
    print()
    print(out_df["explanation7"][exp_num])
    print()
    print(out_df["explanation9"][exp_num])
    print()
    print(out_df["explanation10"][exp_num])
    print()
    print()
    print("-"*80)
    print()
    print()
```

```python
out_df.to_csv("data/asdf.csv")
```
